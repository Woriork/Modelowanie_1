```python
import matplotlib.pyplot as plt


x1_values = []
x2_values = []
y1_values = []
y2_values = []
# Otwarcie pliku i odczytanie danych
with open('data.txt', 'r') as file:
    for line in file:
        # Rozdzielanie wartości na dwie kolumny
        columns = line.strip().split()
        x1_values.append(float(columns[0]))
        y1_values.append(float(columns[1]))
        x2_values.append(float(columns[2]))
        y2_values.append(float(columns[3]))

# Rysowanie wykresu
plt.scatter(x1_values, y1_values, label='Ciało 1')
plt.scatter(x2_values, y2_values, label='Ciało 2')
plt.xlabel('Położenie x')
plt.ylabel('Położenie y')
plt.title('Położenie ciał w przestrzeni')
plt.legend()
plt.grid(True)
plt.show()


```

```python
import matplotlib.pyplot as plt

# Odczyt danych z pliku
t_values = []
e_values = []

with open('data2.txt', 'r') as file:
    for line in file:
        columns = line.strip().split()
        t_values.append(float(columns[0]))
        e_values.append(float(columns[1]))

# Tworzenie wykresu
plt.scatter(t_values, e_values)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Wartość dt')
plt.ylabel('Błąd e')
plt.title('Zależność błędu od wartości kroku czasowego')
plt.grid()
plt.show()

```
