# Analiza Pana Tadeusza #

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

with open('top_words.txt', 'r') as file:
    lines = file.readlines()

# Inicjalizacja list na dane
words = []
frequencies = []

for line in lines:
    word, freq = line.strip().split('\t')
    words.append(word)
    frequencies.append(int(freq))

# Konwersja na logarytmiczne wartoci rangi i czstoci
log_rank = np.log10(range(1, len(words) + 1))
log_freq = np.log10(frequencies)

plt.figure(figsize=(10, 6))
plt.scatter(log_rank, log_freq, color='skyblue', label='Dane')

# Dopasowanie funkcji postaci Zipfa
def zipf(x, a, b):
    return a - b * x

popt, pcov = curve_fit(zipf, log_rank, log_freq)
plt.plot(log_rank, zipf(log_rank, *popt), 'r--', label='Dopasowana linia (Zipf)')

plt.xlabel('Logarytm rangi sowa')
plt.ylabel('Logarytm czstoci wystpowania')
plt.title('Wykres log-log czstoci wystpowania sw z dopasowan lini Zipfa Pan Tadeusz')
plt.legend()
plt.tight_layout()
plt.show()

print("Parametry dopasowanej funkcji Zipfa:")
print("a =", popt[0])
print("b =", popt[1])
```
